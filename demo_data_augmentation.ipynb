```python
import urllib

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras import layers
AUTOTUNE = tf.data.experimental.AUTOTUNE

#import tensorflow_docs as tfdocs
#import tensorflow_docs.plots

import tensorflow_datasets as tfds

import PIL.Image

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (12, 5)

import numpy as np
```

```python
image_path = tf.keras.utils.get_file("cat.jpg", "https://storage.googleapis.com/download.tensorflow.org/example_images/320px-Felis_catus-cat_on_snow.jpg")
PIL.Image.open(image_path)
```

```python
imag_string = tf.io.read_file(image_path)
image = tf.image.decode_jpeg(imag_string, channels=3)
```

```python
def visualisze(original, augmented):
    fig = plt.figure()
    plt.subplot(1,2,1)
    plt.title('Original image')
    plt.imshow(original)
    
    plt.subplot(1,2,2)
    plt.title('Augmented image')
    plt.imshow(augmented)
```

```python
flipped = tf.image.flip_left_right(image)
visualisze(image, flipped)
```

```python
grayscaled = tf.image.rgb_to_grayscale(image)
visualisze(image, tf.squeeze(grayscaled))
```

```python
staurated = tf.image.adjust_saturation(image, 3)
visualisze(image, staurated)
```

```python
bright = tf.image.adjust_brightness(image, 0.35) #the best result is 0.4 on well captured image
visualisze(image, bright)
```

```python
rotated = tf.image.rot90(image)
visualisze(image, rotated)
```

```python
cropped = tf.image.central_crop(image, central_fraction=0.65)
visualisze(image, cropped)
```
